```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.svm import SVC
```

```python
X, y = datasets.make_blobs(n_samples=100, centers=2, random_state=6)
#make_blobs 2ta class yaratadi
#n_samples=100 100ta nuqta yaratadi
#0 va 1 sinfli nuqtalarga ajratiladi
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='winter')
#X[:, 0]; X[:, 1] nuqtalar ornini bildiradi, yani past baland va ong  chap
# c=y agar nuqta 0 bolsa u kok rangda, aks xolda yashil
plt.title("Ma'lumotlar nuqtalari")
plt.show()
```

```python
model = SVC(kernel='linear')  # Linear yassi chiziq bilan ajratadi
model.fit(X, y)
```

```python
def plot_svm_decision_boundary(model, X, y):
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap='winter')

    ax = plt.gca() #biz bergan grafikdagi chegaralarni korsatadi
    xlim = ax.get_xlim() # x oqi chapdan ongacha chegarasi
    ylim = ax.get_ylim() #y oqi chegarasi

    xx = np.linspace(xlim[0], xlim[1]) # chiziq (chapdan, ongga)
    yy = np.linspace(ylim[0], ylim[1]) # tepadan pastga qarab ketgan chiziqdagi nuqta
    YY, XX = np.meshgrid(yy, xx)
    xy = np.vstack([XX.ravel(), YY.ravel()]).T # sanab nuqtalarni juft qiladi

    Z = model.decision_function(xy).reshape(XX.shape)

    # Ajratuvchi chiziq
    ax.contour(XX, YY, Z, colors='k', levels=[-1, 0, 1], linestyles=['--', '-', '--'])
    plt.title("SVM ")
    plt.show()

plot_svm_decision_boundary(model, X, y)
```
